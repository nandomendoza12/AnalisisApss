Proyecto analisis de mercado digital(Apps)

En este proyecto nos centraremos en mostrar como las mayores tiendas de apps (Google Play y AppStore) muestran cuales han sido sus
aplicaciones mas descargadas desde 2018

```python
#Importamos las librerias que vamos a usar en el analisis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Abrimos los archivos 
df1 = pd.read_csv('googleplaystore.csv')
df1
```

```python
print('Imformacion del Data Frame')
df1.info()
print('-------------------------------\n')
print('DESCRIPCION DE DATOS ESTADISTICOS')
df1.describe()
```

```python
# Verificamos si hay valores faltantes como nulos o NaN
if pd.isnull(df1['Rating']).any():
    print("Hay valores faltantes en la columna 'Ranting'")
else:
    print("No hay valores faltantes en la columna 'Ranting'")

#Eliminamos los nulos faltantes en la columna Rating
df1 = df1.dropna(subset=['Rating'])
# Verificamos nuevamente si hay valores faltantes
if pd.isnull(df1['Rating']).any():
    print("Hay valores faltantes en la columna 'Ranting'")
else:
    print("No hay valores faltantes en la columna 'Ranting'")
```

```python
# Verificamos si hay valores faltantes como nulos o NaN
if pd.isnull(df1['Type']).any():
    print("Hay valores faltantes en la columna 'Ranting'")
else:
    print("No hay valores faltantes en la columna 'Ranting'")

#Eliminamos los nulos faltantes en la columna Rating
df1 = df1.dropna(subset=['Type'])
# Verificamos nuevamente si hay valores faltantes
if pd.isnull(df1['Type']).any():
    print("Hay valores faltantes en la columna 'Ranting'")
else:
    print("No hay valores faltantes en la columna 'Ranting'")
```

```python
# Verificamos si hay valores faltantes como nulos o NaN
if pd.isnull(df1['Current Ver']).any():
    print("Hay valores faltantes en la columna 'Current Ver'")
else:
    print("No hay valores faltantes en la columna 'Current Ver'")

#Eliminamos los nulos faltantes en la columna Rating
df1 = df1.dropna(subset=['Current Ver'])
# Verificamos nuevamente si hay valores faltantes
if pd.isnull(df1['Current Ver']).any():
    print("Hay valores faltantes en la columna 'Current Ver'")
else:
    print("No hay valores faltantes en la columna 'Current Ver'")
```

```python
if pd.isnull(df1['Android Ver']).any():
    print("Hay valores faltantes en la columna 'Android Ver'")
else:
    print("No hay valores faltantes en la columna 'Android Ver'")

#Eliminamos los nulos faltantes en la columna Rating
df1 = df1.dropna(subset=['Android Ver'])
# Verificamos nuevamente si hay valores faltantes
if pd.isnull(df1['Android Ver']).any():
    print("Hay valores faltantes en la columna 'Android Ver'")
else:
    print("No hay valores faltantes en la columna 'Android Ver'")
```

```python
# Verificamos si hay valores duplicados
if df1.duplicated().any():
    print("Hay valores duplicados en el DataFrame")
else:
    print("No hay valores duplicados en el DataFrame")
#Mostramos los duplicados
duplicates = df1[df1.duplicated()]
print("Duplicados encontrados:")
print(duplicates)
# Eliminamos los duplicados
df1 = df1.drop_duplicates()
# Verificamos nuevamente si hay valores duplicados
if df1.duplicated().any():
    print("Hay valores duplicados en el DataFrame")
else:
    print("No hay valores duplicados en el DataFrame")
# Mostramos el DataFrame sin duplicados
print("DataFrame sin duplicados:")
print(df1)
```

```python
#Vamos hacer el df1 en un df2 ya que esta mas limpio
df2 = df1.copy()
df2
```

```python
# Verificamos si hay valores nulo y faltantes en el nuevo data frame df2
if df2.isnull().values.any():
    print("Hay valores nulos o faltantes en el DataFrame df2")
else:
    print("No hay valores nulos o faltantes en el DataFrame df2")
# Verificamos si hay valores duplicados en el nuevo data frame df2
if df2.duplicated().any():
    print("Hay valores duplicados en el DataFrame df2")
else:
    print("No hay valores duplicados en el DataFrame df2")
    
```

```python
df2
```

```python
# Con el nuevo data frame df2 vamos a hacer un analisis de los datos
# Vamos a ver la cantidad de aplicaciones por categoria
categoria_counts = df2['Category'].value_counts()
print("Cantidad de aplicaciones por categoría:")
print(categoria_counts)
```

```python
#Convertimos la columna 'Installs' a tipo numérico, eliminando los caracteres no numéricos
df2['Installs'] = df2['Installs'].str.replace('+', '').str.replace(',', '').astype(int)
# Sacamos el top 10 de las aplicaciones mas instaladas y mostramos sus numeros reales de instalacion 
top_installs = df2.nlargest(10, 'Installs')
print("Top 10 aplicaciones más instaladas:")
print(top_installs[['App', 'Installs']])
```

```python
#Pasamos la columna 'Price' a tipo numérico, eliminando el signo de dólar y convirtiendo a float
df2['Price'] = df2['Price'].str.replace('$', '').astype(float)

#Mostramos las aplicaciones que tienen precio de mayor a menor
precio_mayor = df2[df2['Price'] > 0].sort_values(by='Price', ascending=False)
print("Aplicaciones con precio mayor a menor:")
print(precio_mayor[['App', 'Price']])
#Pas

```

```python
# Vamos a verifcar cuales son las aplicaciones mas descargadas por genero
genero_counts = df2['Genres'].value_counts()
print("Cantidad de aplicaciones por género:")
print(genero_counts)
```

```python
# Cuales son las mejores aplicaciones calificadas por genero
mejores_calificadas = df2.nlargest(10, 'Rating')
print("Mejores aplicaciones calificadas:")
print(mejores_calificadas[['App', 'Rating', 'Genres']])
```

```python
#verificcamos si hay nulos en el tamaño de las aplicaciones
""" if pd.isnull(df2['Size']).any():
    print("Hay valores nulos en la columna 'Size'")
else:
    print("No hay valores nulos en la columna 'Size'") """

# Corregimos todos los valores de la columna 'Size' para que sean numéricos
df2['Size'] = df2['Size'].replace('Varies with device', np.nan)
df2['Size'] = df2['Size'].str.replace('M', '').str.replace('k', 'e-3').astype(float)
# Verificamos si hay valores nulos en la columna 'Size'
if pd.isnull(df2['Size']).any():
    print("Hay valores nulos en la columna 'Size'")
else:
    print("No hay valores nulos en la columna 'Size'")
# Mostramos las aplicaciones con el tamaño más grande
tamaño_mayor = df2.nlargest(10, 'Size')
print("Aplicaciones con el tamaño más grande:")
print(tamaño_mayor[['Category', 'App', 'Size']])

#
```

```python
# volvemos  a poner varies with device en la columna de current ver por lo valores Nan
#df2['Current Ver'] = df2['Current Ver'].replace(np.nan, 'Varies with device')

# Vamos a ver las actualizaciones de las aplicaciones y para que version de android  y no mostramos las que tenga varies with device
actualizaciones = df2[df2['Current Ver'] != 'Varies with device'][['App', 'Current Ver']]
print("Actualizaciones de las aplicaciones:")
print(actualizaciones)

print('------------------------------------------------------------------------------\n')

versiones_android = df2[df2['Android Ver'] != 'Varies with device'][['App', 'Android Ver']]
print("Versiones de Android requeridas por las aplicaciones:")
print(versiones_android)



```

```python
# Separamos las apicacion que current ver en varies with device
varies_apps = df2[df2['Current Ver'] == 'Varies with device']
print("Aplicaciones con 'Current Ver' como 'Varies with device':")
print(varies_apps[['App', 'Current Ver']])

print('------------------------------------------------------------------------------\n')
# Mostramos las aplicaciones que tienen 'Varies with device' en 'Android Ver'
varies_android = df2[df2['Android Ver'] == 'Varies with device']
print("Aplicaciones con 'Android Ver' como 'Varies with device':")
print(varies_android[['App', 'Android Ver']])
```

```python
#Analisis en tendencias de instalacion y rating de las aplicaciones
# Agrupamos por categoría y calculamos la media de las instalaciones y el rating
tendencias = df2.groupby('Category').agg({'Installs': 'mean', 'Rating': 'mean'}).reset_index()
print("Tendencias de instalaciones y rating por categoría:")
print(tendencias)
```

```python
# Guardamos el DataFrame limpio en un nuevo archivo CSV ya con los cambios realizados
df2.to_csv('googleplaystore_cleaned.csv', index=False)
# Mostramos el DataFrame limpio
print("DataFrame limpio guardado como 'googleplaystore_cleaned.csv'")
print(df2)
# --- IGNORE ---
```

Visualizaciones de este dataframe df2 

```python
# Un histograma para la distribucion de los precios, tamaños y rating de las aplicaciones 
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.hist(df3['Price'], bins=30, color='blue', alpha=0.5)
plt.title('Distribución de Precios')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.subplot(1, 3, 2)
plt.hist(df3['Size'], bins=30, color='green', alpha=0.5)
plt.title('Distribución de Tamaños')
plt.xlabel('Tamaño')
plt.ylabel('Frecuencia')
plt.subplot(1, 3, 3)
plt.hist(df3['Rating'], bins=30, color='red', alpha=0.5)
plt.title('Distribución de Ratings')
plt.xlabel('Rating')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

```python
#Grafico de barras para la cantidad de aplicaciones por categoria
plt.figure(figsize=(12, 6))
categoria_counts.plot(kind='bar', color='skyblue')
plt.title('Cantidad de Aplicaciones por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Cantidad de Aplicaciones')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```

```python
# Un boxplot para ver la distribucion de los precios, tamaños y rating de las aplicaciones
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.boxplot(df3['Price'], vert=False)
plt.title('Distribución de Precios')
plt.xlabel('Precio')
plt.subplot(1, 3, 2)
plt.boxplot(df3['Size'], vert=False)
plt.title('Distribución de Tamaños')
plt.xlabel('Tamaño')
plt.subplot(1, 3, 3)
plt.boxplot(df3['Rating'], vert=False)
plt.title('Distribución de Ratings')
plt.xlabel('Rating')
plt.tight_layout()
plt.show()
```

```python
# grafico de dispersion para ver la relacion entre el rating y las instalaciones
plt.figure(figsize=(10, 6))
plt.scatter(df3['Rating'], df3['Installs'], alpha=0.5, color='purple')
plt.title('Relación entre Rating e Instalaciones')
plt.xlabel('Rating')
plt.ylabel('Instalaciones')
plt.grid(True)
plt.show()
```

```python
#Graficas de barras para las aplicaciones mas instaladas y mejor calificadas
mejores_calificadas.plot(kind='bar', x='App', y='Installs', color='orange', legend=False)
plt.title('Top 10 Aplicaciones mejores calificadas')
plt.xlabel('Aplicación')
plt.ylabel('Instalaciones')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# grafico de barras para las aplicaciones de mayor tamaño
tamaño_mayor.plot(kind='bar', x='App', y='Size', color='green', legend=False)
plt.title('Top 10 Aplicaciones con mayor tamaño')
plt.xlabel('Aplicación')
plt.ylabel('Tamaño')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
