ANALISAMOS EL SEGUNDO DATAFRAME AHORA DE APLICACION DE APPLE 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('AppleStore.csv')
df
```

```python
descripcion = df.describe()
print(descripcion)


```

```python
df.info()
```

```python
# Vamos hacer una funcion para verificar si hay nulos en el dataframe
def verificar_nulos(df):
    nulos = df.isnull().sum()
    nulos = nulos[nulos > 0]
    if not nulos.empty:
        print("Columnas con valores nulos:")
        print(nulos)
    else:
        print("No hay valores nulos en el DataFrame.")
verificar_nulos(df)
```

```python
# Verificamos si hay duplicados
def verificar_duplicados(df):
    duplicados = df.duplicated().sum()
    if duplicados > 0:
        print(f"Hay {duplicados} filas duplicadas en el DataFrame.")
    else:
        print("No hay filas duplicadas en el DataFrame.")
verificar_duplicados(df)
```

```python
# Vamos a ver cuantas aplicaciones son de cada categoria 
conteo_categorias = df['prime_genre'].value_counts()
print("Cantidad de aplicaciones por categoría:")
print(conteo_categorias)
```

```python
# Ahora vamos a mostrar las aplicaciones con mayor precio a menor y los comparamos con el ranking de las aplicaciones
df.sort_values(by='price', ascending=False, inplace=True)
top_10_precios = df.head(10)
print("Top 10 aplicaciones con mayor precio:")
print(top_10_precios[['track_name', 'price',  'cont_rating']])

```

```python
#Las apps mas populares con el mayor rating
top_10_rating = df.sort_values(by='user_rating', ascending=False).head(10)
print("Top 10 aplicaciones con mayor rating:")
print(top_10_rating[['track_name', 'price', 'cont_rating']])
```

```python
#Vamos a ver el tamaño de las apps
df['size_bytes'] = df['size_bytes'].apply(lambda x: x / (1024 * 1024))  # Convertir a MB
top_10_tamano = df.sort_values(by='size_bytes', ascending=False).head(10)
print("Top 10 aplicaciones con mayor tamaño:")
print(top_10_tamano[['track_name', 'size_bytes', 'cont_rating']])
```

```python
# vamos a ver las app de menor tamaño 
df['size_bytes'] = df['size_bytes'].apply(lambda x: x / (1024 * 1024))  # Convertir a MB
top_10_tamano_menor = df.sort_values(by='size_bytes', ascending=True).head(10)
print("Top 10 aplicaciones con menor tamaño:")
print(top_10_tamano_menor[['track_name', 'size_bytes', 'cont_rating']])
```

```python
# que apps funcionana en mas dispositivos segun user rating
top_10_dispositivos = df.sort_values(by='user_rating', ascending=False).head(10)
print("Top 10 aplicaciones con mayor número de dispositivos:")
print(top_10_dispositivos[['track_name', 'user_rating', 'sup_devices.num']])
```

```python
# Determinamos cuales son las app disponibles en mas idiomas y verificamos la correlacion positiva  con el total de calificaciones
top_10_idiomas = df.sort_values(by='lang.num', ascending=False).head(10)
print("Top 10 aplicaciones con mayor número de idiomas:")
print(top_10_idiomas[['track_name', 'lang.num', 'rating_count_tot']])
print("Correlación entre el número de idiomas y el total de calificaciones:")
print(df[['lang.num', 'rating_count_tot']].corr())
```

```python
# Vamos a ser una relaion donde los usuarios estan contentos con las aplicaciones de mayor tamaño o de menor tamaño
sns.scatterplot(data=df, x='size_bytes', y='user_rating', hue='cont_rating', alpha=0.9)
plt.title('Relación entre tamaño de la aplicación y rating de usuario')
plt.xlabel('Tamaño de la aplicación (MB)')
plt.ylabel('Rating de usuario')
plt.legend(title='Clasificación de contenido')
plt.show()
```

```python
# Rating promedio por Genero
rating_promedio_genero = df.groupby('prime_genre')['user_rating'].mean().sort_values(ascending=False).reset_index()
rating_promedio_genero.columns = ['Género', 'Rating Promedio']
rating_promedio_genero = rating_promedio_genero.head(10)  # Top 10 géneros por rating promedio

print("Rating promedio por género:")
print(rating_promedio_genero)
 
```

```python
#Distribución de precios y tamaños por Genre 
plt.figure(figsize=(14, 7))
sns.boxplot(data=df, x='prime_genre', y='price')
plt.title('Distribución de precios  Género')
plt.xlabel('Género')
plt.ylabel('Precio')
plt.xticks(rotation=45)
plt.show()
```

```python
# Vamos a ver la correlación entre Rating Count y Average User Rating
plt.figure(figsize=(15 ,9))
sns.scatterplot(data=df, x='user_rating', y='rating_count_tot', hue='prime_genre', alpha=0.7)
plt.title('Relación entre Rating Count y Average User Rating')
plt.show()
```

```python
# Estrellas vs. Volumen: Creacion de un gráfico de dispersión de Average User Rating vs. Rating Count. Puedes usar el tamaño de los puntos para representar el Price o el Size.
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='user_rating', y='rating_count_tot', size='price', sizes=(20, 200), hue='prime_genre', alpha=0.6)
plt.title('Relación entre Rating Count y Average User Rating con tamaño de puntos representando el precio')
plt.xlabel('Average User Rating')
plt.ylabel('Rating Count')
plt.legend(title='Género', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Historgramas para la distrubucion de size, price, ratinf_count average useer rating
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
sns.histplot(df['size_bytes'], bins=30, kde=True)
plt.title('Distribución del tamaño de las aplicaciones (MB)')
plt.xlabel('Tamaño (MB)')
plt.ylabel('Frecuencia')
plt.subplot(2, 2, 2)
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribución de precios de las aplicaciones')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.subplot(2, 2, 3)
sns.histplot(df['rating_count_tot'], bins=30, kde=True)
plt.title('Distribución del conteo total de ratings')
plt.xlabel('Conteo total de ratings')
plt.ylabel('Frecuencia')
plt.subplot(2, 2, 4)
sns.histplot(df['user_rating'], bins=30, kde=True)
plt.title('Distribución del rating promedio de las aplicaciones')
plt.xlabel('Rating promedio')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

```python
#Grafico de barras para la cantidad de aplicaciones por género
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='prime_genre', order=df['prime_genre'].value_counts().index)
plt.title('Cantidad de aplicaciones por género')
plt.xlabel('Género')
plt.ylabel('Cantidad de aplicaciones')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plots para la comparacion de distribucion numerica size, price , rating coun entre géneros
plt.figure(figsize=(14, 7))
sns.boxplot(data=df, x='prime_genre', y='size_bytes')
plt.title('Distribución del tamaño de las aplicaciones por género')
plt.xlabel('Género')
plt.ylabel('Tamaño (MB)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plots para ver las correlaciones entre variable size, price, rating_count_tot y user_rating
plt.figure(figsize=(15, 10))
sns.scatterplot(data=df, x='size_bytes', y='user_rating', hue='prime_genre', alpha=0.7)
plt.title('Relación entre tamaño de la aplicación y rating de usuario')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.scatterplot(data=df, x='price', y='user_rating', hue='prime_genre', alpha=0.7)
plt.title('Relación entre precio de la aplicación y rating de usuario')
plt.show()
```

```python
# Heatmaps de correlación de variables numericas 
plt.figure(figsize=(12, 8))
correlation_matrix = df[['size_bytes', 'price', 'rating_count_tot', 'user_rating']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de correlación entre variables numéricas')
plt.show()
```
